```python
plt.hist(df['salary'], bins=20, edgecolor='black')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Salary Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x='job_title', y='salary', data=df, errorbar=None)
plt.xticks(rotation=45, ha='right')
plt.xlabel('Job Title')
plt.ylabel('Average Salary')
plt.title('Average Salary by Job Title')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='experience_level', y='salary', data=df, errorbar=None)
plt.xlabel('Experience Level')
plt.ylabel('Average Salary')
plt.title('Average Salary by Experience Level')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='employment_type', y='salary', data=df)
plt.xlabel('Empolyment type')
plt.ylabel('Salary')
plt.title('Salary Distribution by Education Level')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='experience_level', data=df)
plt.xlabel('Experience Level')
plt.ylabel('Count')
plt.title('Count of Each Experience Level')
plt.show()
```

```python
correlation_matrix = df.corr(numeric_only=True)

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
